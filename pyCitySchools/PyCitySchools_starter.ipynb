```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_file = "Resources/schools_complete.csv"
student_file = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_file)
school_data_df=pd.DataFrame(school_data)

student_data = pd.read_csv(student_file)
student_data_df=pd.DataFrame(student_data)

# Combine the data into a single dataset.  
school_data_combined_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name"])
school_data_combined_df.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools=school_data_df["school_name"].count()

# Calculate the total number of students
total_students=student_data_df["student_name"].count()

# Calculate the total budget
total_budget=school_data_df["budget"].sum()

# Calculate the average maths score
average_maths_score=student_data_df["maths_score"].mean()

# Calculate the average reading score
average_reading_score=student_data_df["reading_score"].mean()

# Calculate the percentage of students with a passing maths score (50 or greater)
passing_maths=student_data_df.loc[student_data_df["maths_score"]>=50]
pass_percentage_maths=(passing_maths["maths_score"].count()/total_students)*100

# Calculate the percentage of students with a passing reading score (50 or greater)
passing_reading=student_data_df.loc[student_data_df["reading_score"]>=50]
pass_percentage_reading=(passing_reading["reading_score"].count()/total_students)*100

# Calculate the percentage of students who passed maths and reading (% Overall Passing)
total_passing_percentage=(school_data_combined_df[(school_data_combined_df["maths_score"]>=50)&(school_data_combined_df["reading_score"]>=50)]["Student ID"].count()/total_students)*100
total_passing_percentage


# Create a dataframe to hold the above results
local_government_area_summary=pd.DataFrame({"Total Schools":[total_schools],"Total Students":[total_students],"Total Budget":[total_budget],"Average Maths Score":[average_maths_score],"Average Reading Score":[average_reading_score],"Maths Passing Percentage":[pass_percentage_maths],"Reading Passing Percentage":[pass_percentage_reading],"Total Pass Percentage":[total_passing_percentage]})
local_government_area_summary.head()
# Optional: give the displayed data cleaner formatting
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create an overview table that summarises key metrics about each school, including:
school_name_type = school_data.set_index(["school_name"])["type"]
school_name_type
# School Name
# School Type
# school_name_type=school_data_combined_df.groupby(["school_name","type"])
# school_name_type_df=pd.DataFrame(school_name_type)
# # Total Students
total_students_in_school=school_data_combined_df["school_name"].value_counts()
total_students_in_school

# # Total School Budget
school_budget=school_data_combined_df["budget"].mean()
school_budget

# # Per Student Budget
per_student_budget=school_budget/total_students_in_school
per_student_budget

# # Average Maths Score
average_maths_score_per_school=school_data_combined_df["maths_score"].mean()
average_maths_score_per_school

# # Average Reading Score
average_reading_score_per_school=school_data_combined_df["reading_score"].mean()
average_reading_score_per_school

# # % Passing Maths
passing_maths_per_school=school_data_combined_df[(school_data_combined_df["maths_score"]>=50)]
passing_maths_percentage_per_school=passing_maths_per_school.groupby(["school_name"]).count()["student_name"]/total_students_in_school*100
passing_maths_percentage_per_school

# # % Passing Reading
passing_reading_per_school=school_data_combined_df[(school_data_combined_df["reading_score"]>=50)]
passing_reading_percentage_per_school=passing_reading_per_school.groupby(["school_name"]).count()["student_name"]/total_students_in_school*100
passing_reading_percentage_per_school

# # % Overall Passing (The percentage of students that passed maths and reading.)
passing_reading_and_maths_percentage=(school_data_combined_df[(school_data_combined_df["reading_score"]>=50)&(school_data_combined_df["maths_score"]>=50)].groupby("school_name")["Student ID"].count()/total_students_in_school)*100
passing_reading_and_maths_percentage

# # Create a dataframe to hold the above results
school_summary=pd.DataFrame({"School Type":school_name_type,"Total Students":total_students_in_school,"Total School Budget":school_budget,"Per Student Budget":per_student_budget,"Average Maths Score":average_maths_score_per_school,"Average Reading Score":average_reading_score_per_school,"% Passing Maths":passing_maths_percentage_per_school,"% Passing Reading":passing_reading_percentage_per_school,"% Overall Passing":passing_reading_and_maths_percentage})
school_summary
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
year_nine = student_data_df.loc[student_data["year"] == "9"]
year_nine_maths=student_data_df.groupby("school_name")["maths_score"].mean()
year_nine_maths

year_ten= student_data_df.loc[student_data["year"] == "10"]
year_ten_maths=student_data_df.groupby("school_name")["maths_score"].mean()
year_ten_maths

year_eleven = student_data_df.loc[student_data["year"] == "11"]
year_eleven_maths=student_data_df.groupby('school_name')["maths_score"].mean()
year_eleven_maths

year_twelve = student_data_df.loc[student_data["year"] == "12"]
year_twelve_maths= student_data_df.groupby('school_name')["maths_score"].mean()
year_twelve_maths
maths_score_per_year = pd.DataFrame({"9": year_nine_maths,      
                                    "10": year_ten_maths,
                                    "11": year_eleven_maths,
                                    "12": year_twelve_maths})
maths_score_per_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 584, 630, 645, 680]
group_name = ["<$585", "$585-630", "$630-645", "$645-680"]

school_summary["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, bins, labels = group_name)
school_summary

ave_maths_scores= school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
ave_maths_scores

ave_reading_scores=school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
ave_reading_scores

percent_passing_maths=school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
percent_passing_maths

percent_passing_reading=school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
percent_passing_reading

overall_passing_rate=(percent_passing_maths+percent_passing_reading)/2
overall_passing_rate

performance_by_school_spending = pd.DataFrame({"Average Math Score":ave_maths_scores,
                                    "Average Reading Score":ave_reading_scores,
                                    "% Passing Math":percent_passing_maths,
                                    "% Passing Reading":percent_passing_reading,
                                    "% Overall Passing Rate": overall_passing_rate})
performance_by_school_spending
```
